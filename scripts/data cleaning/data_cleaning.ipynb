```python
cleaned_data = data.drop(columns=["Shipping Address.1", "Shipping Address Duplicate"])
```

```python
cleaned_data['Phone Number'].sample(15,random_state=2)
```

```python
def standardize_phone_number(phone):
    match = re.match(r'.*?(\d{3}).*?(\d{3}).*?(\d{4})', phone)
    if match:
        return f"+1 ({match.group(1)}) {match.group(2)}-{match.group(3)}"
    return phone
```

```python
cleaned_data['Phone Number'] = cleaned_data['Phone Number'].apply(standardize_phone_number)
```
